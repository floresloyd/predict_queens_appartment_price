## I. Train Model on Full Dataset

```python
from sklearn.ensemble import BaggingRegressor
from sklearn.tree import DecisionTreeRegressor
import pandas as pd

df = pd.read_csv("housing_clean_imputed_data.csv")
df
```

```python
print("Splitting X and y ...")
# Combine training and validation sets
X = df.drop('sale_price', axis=1)
y = df['sale_price']

print("Training Best Model: Bagged Decision Trees")
# Best hyperparameters
best_params = {'max_depth': None, 'min_samples_leaf': 1, 'min_samples_split': 2, 'n_estimators': 300}

# Initialize the Bagging Regressor with the best hyperparameters
best_model = BaggingRegressor(
    base_estimator=DecisionTreeRegressor(max_depth=best_params['max_depth'], 
                                         min_samples_split=best_params['min_samples_split'], 
                                         min_samples_leaf=best_params['min_samples_leaf']),
    n_estimators=best_params['n_estimators'], 
    random_state=42
)

print("Fitting model on entire dataset ...")
# Train the model on the entire dataset
best_model.fit(X, y)

print("Model training complete...")
```

```python
import numpy as np

# Example function to predict price based on input features
def predict_price(
    sale_year, zone, cats_allowed, dogs_allowed, coop_condo, garage_exists,
    num_bedrooms, num_floors_in_building, num_full_bathrooms, num_half_bathrooms,
    num_total_rooms, dining_room_type, fuel_type, kitchen_type, community_district_num,
    approx_year_built, sq_footage, walk_score, maintenance_cost, parking_charges
):
    # Combine the input features into a single array
    input_features = np.array([
        sale_year, zone, cats_allowed, dogs_allowed, coop_condo, garage_exists,
        num_bedrooms, num_floors_in_building, num_full_bathrooms, num_half_bathrooms,
        num_total_rooms, dining_room_type, fuel_type, kitchen_type, community_district_num,
        approx_year_built, sq_footage, walk_score, maintenance_cost, parking_charges
    ]).reshape(1, -1)
    
    # Check if the model is expecting a different set of features
    if hasattr(best_model, 'feature_names_in_'):
        expected_features = best_model.feature_names_in_
        print(f"Model expects features: {expected_features}")
    
    # Predict the price
    predicted_price = best_model.predict(input_features)
    
    # Generate the feature description
    feature_description = (
        f"Sale year: {sale_year}, Zone: {zone}, Cats allowed: {cats_allowed}, "
        f"Dogs allowed: {dogs_allowed}, Coop/Condo: {coop_condo}, Garage exists: {garage_exists}, "
        f"Bedrooms: {num_bedrooms}, Floors in building: {num_floors_in_building}, "
        f"Full bathrooms: {num_full_bathrooms}, Half bathrooms: {num_half_bathrooms}, "
        f"Total rooms: {num_total_rooms}, Dining room type: {dining_room_type}, Fuel type: {fuel_type}, "
        f"Kitchen type: {kitchen_type}, Community district number: {community_district_num}, "
        f"Approx. year built: {approx_year_built}, Square footage: {sq_footage}, "
        f"Walk score: {walk_score}, Maintenance cost: {maintenance_cost}, Parking charges: {parking_charges}"
    )
    
    return predicted_price[0], feature_description

# Example usage with the provided sample values
sale_year = 2017  # (2016, 2017)
zone = 2  # (1 - 9)
cats_allowed = 1  # (0,1)
dogs_allowed = 1  # (0,1)
coop_condo = 2  # (1:Coop, 2:Condo)
garage_exists = 0  # (0,1)
num_bedrooms = 1  # (1,..,n)
num_floors_in_building = 5  # (1,..,n)
num_full_bathrooms = 1  # (1,..,n)
num_half_bathrooms = 0  # (1,..,n)
num_total_rooms = 1  # (1,...,n)
dining_room_type = 1  # (1:none, 2:dining area, 3:combo, 4:other, 5:formal)
fuel_type = 5  # (1:oil, 2:none, 3:gas, 4:other, 5:electric)
kitchen_type = 0  # (1: none, 2:efficiency, 3:other, 4:eatin, 5:combo)
community_district_num = 26
approx_year_built = 1950  # (19xx - 20xx)
sq_footage = 1200
walk_score = 99
maintenance_cost = 500
parking_charges = 0

# Assuming best_model is already defined and trained
predicted_price, feature_description = predict_price(
    sale_year, zone, cats_allowed, dogs_allowed, coop_condo, garage_exists,
    num_bedrooms, num_floors_in_building, num_full_bathrooms, num_half_bathrooms,
    num_total_rooms, dining_room_type, fuel_type, kitchen_type, community_district_num,
    approx_year_built, sq_footage, walk_score, maintenance_cost, parking_charges
)
formatted_price = f"${predicted_price:,.0f}"
print(f"Predicted Price: {formatted_price}")
print(f"Feature Description: {feature_description}")
```
